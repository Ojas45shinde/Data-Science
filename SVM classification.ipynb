```python
x=digits.data
print(x.shape)
```

```python
y=digits.target
print("The shape of target variable is:",y.shape)
```

```python
from sklearn.decomposition import PCA
from mpl_toolkits.mplot3d import axes3d
```

```python
p=PCA(n_components=3)
x_pca=p.fit_transform(x)
```

```python
fig=plt.figure(figsize=(6,6))
ax=fig.add_subplot(111,projection="3d")

scatter=ax.scatter(x_pca[:,0],x_pca[:,1],x_pca[:,2],c=y,cmap="summer")
legend=ax.legend(*scatter.legend_elements(),title="digits")
ax.add_artist(legend)

ax.set_xlabel("PCA 1")
ax.set_ylabel("PCA 2")
ax.set_zlabel("PCA 3")
plt.show()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)

print("The length of training set: ",len(x_train))

print("The length of testing set:",len(x_test))
```

```python
model.fit(x_train,y_train)
print("The score of the SVM is:",model.score(x_test,y_test))
```
